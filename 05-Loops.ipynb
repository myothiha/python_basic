# Loop over a sequence

```python
students = ['Mg Kaung', 'Moe Theint', 'Aye Nyein', 'Tata', 'Ko Tin Maung', 'Mg Mg']
```

```python
for name in students:
    print(f"Hello {name}")
```

```python
# Dictionary
students = {'Mg Kaung': 30, 'Aye Nyein': 20, 'Moe Theint': 21}
```

```python
# loop over key.
for name in students:
    print(name)
```

```python
# loop over values.
for age in students.values():
    print(age)
```

```python
# loop over both key and values.
for name, age in students.items():
    print(f"Name: {name}")
    print(f"Age: {age}")
```
